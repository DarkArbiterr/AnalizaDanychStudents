```python
import random
import numpy as np
import pingouin as pg
from scipy.stats import ttest_ind, ttest_rel, f_oneway
from scipy import stats
hom = pg.homoscedasticity(data=dane, dv='G3', group='age')
print(hom)
print("\n")
#raczej użyc anovy, bo kruska jest do nieparametrycznych
aov = pg.anova(data=dane, dv="G3", between="age")
print(aov)
print("\n")
krus = pg.kruskal(data=dane, dv='G3', between='age')
print(krus)
```
